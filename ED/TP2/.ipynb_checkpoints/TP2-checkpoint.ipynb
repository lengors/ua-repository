# RAVEN Dataset Analysis

### Modules

In this section we can see all modules used in this work.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA, KernelPCA
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```


### Load gender information

In this section we read data from about gender information.

```python
# set columns and classes
gicolumns = [ 'Indiv', 'Gender' ]
classes = [ 'Feminino', 'Masculino' ]

# load and filter gender info
gender_info = pd.read_csv(os.path.join('RAVEN', 'data.csv'), header = None, engine = 'python')
gender_info = gender_info[gender_info[1].isin(classes)]

# calculate ratios of correct answers
gender_info['training'] = gender_info[2] / (gender_info[2] + gender_info[3])
gender_info['testing'] = gender_info[4] / (gender_info[4] + gender_info[5])

# rename and select gender and indiv columns
gender_info.rename(columns = dict(enumerate(gicolumns)), inplace = True)
gender_info = gender_info[gicolumns + [ 'training', 'testing' ]]

# show gender info
gender_info
```


### Load overall energy ratios data

In this section we load the overall energy ratios data.

```python
# load overall energy ratios and merge with gnender info
overall_energy_ratios = pd.read_excel(os.path.join('RAVEN', 'overall_energy_ratios.xlsx'))
overall_immersion = pd.read_excel(os.path.join('RAVEN', 'overall_immersion.xlsx'))
overall_immersion.columns = [ 'immersion_{}'.format('_'.join(column.split(' '))) for column in overall_immersion.columns ]

overall_energy_ratios = pd.concat([ overall_energy_ratios, overall_immersion ], axis = 1)

# overall_energy_ratios = pd.read_excel(os.path.join('RAVEN', 'Overall_P100.xlsx'))
# overall_energy_ratios = pd.merge(overall_energy_ratios, overall_P100, on = 'Indiv')

# overall_energy_ratios = pd.read_excel(os.path.join('RAVEN', 'Overall_P300.xlsx'))
# overall_energy_ratios = pd.merge(overall_energy_ratios, overall_P300, on = 'Indiv')

overall_energy_ratios = pd.merge(overall_energy_ratios, gender_info, on = 'Indiv')

# select columns and rows
overall_energy_ratios = overall_energy_ratios[[ feature for feature in overall_energy_ratios.columns if 'correct' not in feature ]]
overall_energy_ratios.dropna(inplace = True)
overall_energy_ratios = overall_energy_ratios[overall_energy_ratios[[ feature for feature in overall_energy_ratios.columns if feature not in gender_info.columns ]].sum(axis = 1) > 0]

# display overall energy ratios
overall_energy_ratios
```

```python
y = overall_energy_ratios['Gender'].values
xs = list()
for phase in [ 'training', 'testing' ]:
    for value in [ 'display', 'solutions' ]:
        columns = [ column for column in overall_energy_ratios.columns if phase in column and value in column ]
        for column in columns:
            X = overall_energy_ratios[[ column, phase ]].values
            xs.append(X)
                
plt.scatter(xs[0][y == 'Feminino', 0], xs[0][y == 'Feminino', 1], color='red', marker='^', alpha=0.8)
plt.scatter(xs[0][y == 'Masculino', 0], xs[0][y == 'Masculino', 1], color='blue', marker='o', alpha=0.8)
plt.grid()
# plt.xlim([, 7])
# plt.ylim([-1, 6])
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):
# setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    #plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
    np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
            alpha=0.8, c=cmap(idx),marker=markers[idx], label=cl)
```

```python
from sklearn.linear_model import SGDClassifier, Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.cluster import KMeans

classifier = KMeans(n_clusters = 2, init = 'random', n_init = 10, max_iter = 300, tol = 1e-04, random_state = 0)
# classifier = MLPClassifier(activation = 'tanh', hidden_layer_sizes = (10, 5), alpha = 0.01, max_iter = 5000)
# classifier = SGDClassifier(loss = 'log', max_iter = 5)
# classifier = LinearSVC(C = 1.0, max_iter = 10001, tol = 1e-05, verbose = 0)
# classifier = Perceptron(penalty = None, alpha = 0.0001, fit_intercept = True, max_iter = 10000, tol = None, 
#                eta0 = 0.1, n_jobs = 1, random_state = 0, class_weight = None, warm_start = False)
# classifier = SVC(C = 1.0, kernel = 'rbf', max_iter = 10000, tol = 1e-05, verbose = 0)

classifier
```

```python
yset = set(y)
yset = dict([ (value, i) for i, value in enumerate(yset) ])
ry = [ yset[value] for value in y ]

results = classifier.fit(xs[0], ry)
plot_decision_regions(xs[0], ry, classifier=results)
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```
